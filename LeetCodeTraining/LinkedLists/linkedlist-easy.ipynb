### Definition for singly-linked list.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def __repr__(self):  # nice debug printing
        return f"ListNode({self.val})"
```

### Helpers for local testing

```python
def build_list(values):
    """Return (head, nodes) where nodes preserves object identity order."""
    dummy = ListNode()
    cur = dummy
    nodes = []
    for v in values:
        node = ListNode(v)
        nodes.append(node)
        cur.next = node
        cur = node
    return dummy.next, nodes


def to_list(head):
    out = []
    while head:
        out.append(head.val)
        head = head.next
    return out
```

```python
from image_toggle_gallery import ImageToggleGallery, make_img_html

image_sets = {
    "Page 1": make_img_html(["./docs/0021/0021-02.png"]),
    "Page 2": make_img_html(["./docs/0021/0021-03.png"]),
    "Page 3": make_img_html(["./docs/0021/0021-04.png"]),
}

gallery = ImageToggleGallery(image_sets)
gallery.display()
```

```python
class Solution:
    # Time: O(n + m)
    # Space: O(1)
    def mergeTwoLists(self, list1: ListNode | None, list2: ListNode | None) -> ListNode | None:
        prehead = ListNode(-1)
        prev = prehead

        while list1 and list2:
            if list1.val <= list2.val:
                prev.next = list1
                list1 = list1.next
            else:
                prev.next = list2
                list2 = list2.next

            prev = prev.next

        prev.next = list1 if list1 else list2
        return prehead.next
```

```python
s = Solution()

# 1) Standard example
h1, _ = build_list([1, 2, 4])
h2, _ = build_list([1, 3, 4])
assert to_list(s.mergeTwoLists(h1, h2)) == [1, 1, 2, 3, 4, 4]

# 2) One list empty
h1, _ = build_list([])
h2, _ = build_list([0])
assert to_list(s.mergeTwoLists(h1, h2)) == [0]

# 3) Both empty
assert to_list(s.mergeTwoLists(None, None)) == []

# 4) Different lengths, negatives, duplicates
a = [-10, -10, -9, -4, 1, 6, 6]
b = [-7, -3, 0, 7, 8]
h1, _ = build_list(a)
h2, _ = build_list(b)
assert to_list(s.mergeTwoLists(h1, h2)) == sorted(a + b)

# 5) Stability on equal values (uses <= so nodes from list1 should come first)
h1, n1 = build_list([1, 1, 2])
h2, n2 = build_list([1, 3])
merged = s.mergeTwoLists(h1, h2)
assert merged is n1[0] and merged.next is n1[1], "Equal values should prefer nodes from list1"


# 6) No new node allocation (all nodes in result come from inputs)
def collect_nodes(head):
    out = []
    while head:
        out.append(head)
        head = head.next
    return out


h1, n1 = build_list([1, 2, 4])
h2, n2 = build_list([1, 3, 4])
merged = s.mergeTwoLists(h1, h2)
pool = set(map(id, n1 + n2))
for node in collect_nodes(merged):
    assert id(node) in pool, "Merged list should reuse input nodes (no new nodes created)"

print("✅ All mergeTwoLists assertions passed")
```

```python
class Solution:
    # Time: O(n)
    # Space: O(1)
    def deleteDuplicates(self, head: ListNode | None) -> ListNode | None:
        current = head
        # while current is not None and current.next is not None:
        while current and current.next:
            if current.val == current.next.val:
                current.next = current.next.next
            else:
                current = current.next
        return head
```

```python
def assert_identity_chain(head, expected_nodes):
    """Assert that following `next` pointers from head matches expected node objects exactly."""
    cur = head
    for i, node in enumerate(expected_nodes):
        assert cur is node, f"Node identity mismatch at position {i}"
        cur = cur.next
    assert cur is None, "Result list longer than expected"


s = Solution()

# 1) empty list
head, nodes = build_list([])
res = s.deleteDuplicates(head)
assert to_list(res) == []
assert_identity_chain(res, [])

# 2) single node
head, nodes = build_list([1])
res = s.deleteDuplicates(head)
assert to_list(res) == [1]
assert res is nodes[0]
assert_identity_chain(res, [nodes[0]])

# 3) all duplicates
head, nodes = build_list([1, 1, 1])
res = s.deleteDuplicates(head)
assert to_list(res) == [1]
assert res is nodes[0]
assert_identity_chain(res, [nodes[0]])

# 4) duplicates at head
head, nodes = build_list([1, 1, 2])
res = s.deleteDuplicates(head)
assert to_list(res) == [1, 2]
assert res is nodes[0]
assert_identity_chain(res, [nodes[0], nodes[2]])

# 5) duplicates in middle and end
head, nodes = build_list([1, 1, 2, 3, 3])
res = s.deleteDuplicates(head)
assert to_list(res) == [1, 2, 3]
# kept nodes: first 1 (idx 0), 2 (idx 2), first 3 (idx 3)
assert_identity_chain(res, [nodes[0], nodes[2], nodes[3]])

# 6) no duplicates
head, nodes = build_list([1, 2, 3])
res = s.deleteDuplicates(head)
assert to_list(res) == [1, 2, 3]
assert_identity_chain(res, [nodes[0], nodes[1], nodes[2]])

# 7) long runs of duplicates
vals = [1, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5]
head, nodes = build_list(vals)
res = s.deleteDuplicates(head)
assert to_list(res) == [1, 2, 3, 4, 5]
# kept indices: 0 (1), 1 (first 2), 4 (first 3), 6 (first 4), 10 (5)
assert_identity_chain(res, [nodes[i] for i in [0, 1, 4, 6, 10]])

# 8) negatives (still sorted)
head, nodes = build_list([-1, -1, 0, 0, 0, 1])
res = s.deleteDuplicates(head)
assert to_list(res) == [-1, 0, 1]
# kept indices: 0 (-1), 2 (first 0), 5 (1)
assert_identity_chain(res, [nodes[i] for i in [0, 2, 5]])

# 9) unsorted case (adjacent dups only)
head, nodes = build_list([1, 2, 1])
res = s.deleteDuplicates(head)
assert to_list(res) == [1, 2, 1]
assert_identity_chain(res, [nodes[0], nodes[1], nodes[2]])

print("All deleteDuplicates tests passed ✅")
```

### Solution: 141. Linked List Cycle

```python
from image_toggle_gallery import ImageToggleGallery, make_img_html

image_sets = {
    "Page 1": make_img_html(["../Patterns/Fast and Slow Pointers.png"]),
    "Page 2": make_img_html(["./docs/0141/0141-02.png"]),
    "Page 3": make_img_html(["./docs/0141/0141-03.png"]),
}

gallery = ImageToggleGallery(image_sets)
gallery.display()
```

```python
class Solution:
    # Time: O(n)
    # Space: O(1)
    def hasCycle(self, head: ListNode | None) -> bool:
        if not head:
            return False

        slow = head
        fast = head.next

        while slow != fast:
            if not fast or not fast.next:
                return False

            slow = slow.next
            fast = fast.next.next

        return True
```

```python
def link_tail_to(nodes, pos):
    """Make a cycle by linking tail to nodes[pos] (0-based). pos < 0 -> no-op."""
    if nodes and pos >= 0:
        nodes[-1].next = nodes[pos]


s = Solution()

# 1) empty list
head, nodes = build_list([])
assert s.hasCycle(head) is False

# 2) single node, no cycle
head, nodes = build_list([1])
assert s.hasCycle(head) is False

# 3) single node, self-cycle
head, nodes = build_list([1])
nodes[0].next = nodes[0]
assert s.hasCycle(head) is True

# 4) two nodes, no cycle
head, nodes = build_list([1, 2])
assert s.hasCycle(head) is False

# 5) two nodes, cycle tail -> head
head, nodes = build_list([1, 2])
link_tail_to(nodes, 0)
assert s.hasCycle(head) is True

# 6) longer list, no cycle
head, nodes = build_list([1, 2, 3, 4, 5])
assert s.hasCycle(head) is False

# 7) longer list, cycle tail -> head (pos = 0)
head, nodes = build_list([1, 2, 3, 4, 5])
link_tail_to(nodes, 0)
assert s.hasCycle(head) is True

# 8) longer list, cycle tail -> middle (pos = 2)
head, nodes = build_list([1, 2, 3, 4, 5, 6])
link_tail_to(nodes, 2)
assert s.hasCycle(head) is True

# 9) cycle of length 2 (tail -> previous node)
head, nodes = build_list([10, 20, 30, 40])
nodes[-1].next = nodes[-2]
assert s.hasCycle(head) is True

print("All hasCycle tests passed ✅")
```

### Solution: 206. Reverse Linked List

```python
from image_toggle_gallery import ImageToggleGallery, make_img_html

image_sets = {
    "Page 1": make_img_html(["../Patterns/In-Place Reversal of Linked List.png"]),
    "Page 2": make_img_html(["./docs/0206/0206-02.png"]),
    "Page 3": make_img_html(["./docs/0206/0206-03.png"]),
}

gallery = ImageToggleGallery(image_sets)
gallery.display()
```

```python
class Solution:
    # Time: O(n)
    # Space: O(1)
    def reverseList(self, head: ListNode | None) -> ListNode | None:
        prev = None
        curr = head

        while curr:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp

        return prev
```

```python
def assert_identity_chain(head, expected_nodes):
    """Assert that following next-pointers equals exactly these node objects."""
    cur = head
    for i, node in enumerate(expected_nodes):
        assert cur is node, f"Node identity mismatch at position {i}"
        cur = cur.next
    assert cur is None, "Result list longer than expected"


s = Solution()

# 1) empty list
head, nodes = build_list([])
res = s.reverseList(head)
assert res is None
assert to_list(res) == []
assert_identity_chain(res, [])

# 2) single node
head, nodes = build_list([1])
res = s.reverseList(head)
assert res is nodes[0]
assert to_list(res) == [1]
assert_identity_chain(res, [nodes[0]])

# 3) two nodes
head, nodes = build_list([1, 2])
res = s.reverseList(head)
assert to_list(res) == [2, 1]
assert_identity_chain(res, [nodes[1], nodes[0]])
assert nodes[0].next is None  # former head is now tail

# 4) three nodes
head, nodes = build_list([1, 2, 3])
res = s.reverseList(head)
assert to_list(res) == [3, 2, 1]
assert_identity_chain(res, [nodes[2], nodes[1], nodes[0]])

# 5) duplicates
head, nodes = build_list([1, 1, 2, 3, 3])
res = s.reverseList(head)
assert to_list(res) == [3, 3, 2, 1, 1]
assert_identity_chain(res, [nodes[4], nodes[3], nodes[2], nodes[1], nodes[0]])

# 6) negatives / zeros
head, nodes = build_list([0, -1, -2])
res = s.reverseList(head)
assert to_list(res) == [-2, -1, 0]
assert_identity_chain(res, [nodes[2], nodes[1], nodes[0]])

# 7) longer list
vals = [1, 2, 3, 4, 5, 6]
head, nodes = build_list(vals)
res = s.reverseList(head)
assert to_list(res) == list(reversed(vals))
assert_identity_chain(res, list(reversed(nodes)))

# 8) double-reverse returns original order and identities
head, nodes = build_list([10, 20, 30, 40])
res1 = s.reverseList(head)
res2 = s.reverseList(res1)
assert to_list(res2) == [10, 20, 30, 40]
assert_identity_chain(res2, [nodes[0], nodes[1], nodes[2], nodes[3]])

print("All reverseList tests passed ✅")
```
