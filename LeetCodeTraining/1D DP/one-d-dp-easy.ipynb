### 70. Climbing Stairs

<div><div class="elfjS" data-track-load="description_content"><p>You are climbing a staircase. It takes <code>n</code> steps to reach the top.</p>

<p>Each time you can either climb <code>1</code> or <code>2</code> steps. In how many distinct ways can you climb to the top?</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>

<pre><strong>Input:</strong> n = 2
<strong>Output:</strong> 2
<strong>Explanation:</strong> There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps
</pre>

<p><strong class="example">Example 2:</strong></p>

<pre><strong>Input:</strong> n = 3
<strong>Output:</strong> 3
<strong>Explanation:</strong> There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= n &lt;= 45</code></li>
</ul>

```python
class Solution:
    # Time: O(n)
    # Space: O(n)
    def climbStairs(self, n: int) -> int:
        if n == 1:
            return 1
        dp = [0 for _ in range(n + 1)]
        dp[1] = 1
        dp[2] = 2
        for i in range(3, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2]
        return dp[n]
```

```python
s = Solution()


# quick Fibonacci helper for the property check
def fib(k: int) -> int:
    a, b = 0, 1
    for _ in range(k):
        a, b = b, a + b
    return a


# Exact values
cases = [
    (1, 1),
    (2, 2),
    (3, 3),
    (4, 5),
    (5, 8),
    (10, 89),
    (20, 10946),
    (30, 1346269),
    (45, 1836311903),  # LeetCode max typical constraint
]
for n, expected in cases:
    got = s.climbStairs(n)
    assert got == expected, f"n={n}: expected {expected}, got {got}"

# Property: climbStairs(n) == fib(n+1)
for n in range(1, 46):
    assert s.climbStairs(n) == fib(n + 1), f"n={n}: not equal to fib(n+1)"

# Monotonicity: strictly increasing for n>=1
prev = s.climbStairs(1)
for n in range(2, 15):
    cur = s.climbStairs(n)
    assert cur > prev, f"n={n}: sequence not increasing (prev={prev}, cur={cur})"
    prev = cur

print("âœ… climbStairs assertions passed")
```
