### Modified Binary Search
<img src="../Patterns/Modified Binary Search.png" width="90%" height="auto" style="display: block; margin: auto;">

```python
class Solution:
    # Time: O(log n)
    # Space: O(1)
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x

        left, right = 2, x // 2

        while left <= right:
            pivot = left + (right - left) // 2
            num = pivot * pivot
            if num > x:
                right = pivot - 1
            elif num < x:
                left = pivot + 1
            else:
                return pivot

        return right
```

```python
s = Solution()


def check(x, expected=None):
    y = s.mySqrt(x)
    if expected is not None:
        assert y == expected, f"x={x}: expected {expected}, got {y}"
    # floor-sqrt invariant
    assert y * y <= x < (y + 1) * (y + 1), f"x={x}: invariant failed, y={y}"


# Edge cases
check(0, 0)
check(1, 1)
check(2, 1)
check(3, 1)

# Perfect squares
check(4, 2)
check(9, 3)
check(16, 4)
check(10**16, 10**8)

# Non-squares near squares
check(8, 2)  # between 2^2 and 3^2
check(15, 3)  # between 3^2 and 4^2
check(24, 4)  # between 4^2 and 5^2

# Classic LeetCode 32-bit edge cases
check(2147395599, 46339)  # just below 46340^2
check(2147483647, 46340)  # INT_MAX

print("âœ… mySqrt assertions passed")
```
